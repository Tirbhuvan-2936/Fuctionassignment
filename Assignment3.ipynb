6. Explain the concept of generators in Python and how they are defined.



A generator is a special type of iterator that yields items one at a time instead of storing the entire sequence in memory.

It is memory-efficient and suitable for large datasets or streams of data.

Generators produce values on the fly and are lazy, meaning they generate items only when requested.

1. Using a function with yield

Use the yield keyword instead of return.

Each call to next() on the generator resumes from where it left off.

Example:

```python
def my_generator(n):
    for i in range(n):
        yield i

gen = my_generator(5)

print(next(gen))
print(next(gen))
print(next(gen))

for val in gen:
    print(val)
```

9. Explain the purpose and usage of the `map()` function in Python.

Purpose of map()

The map() function is used to apply a given function to all items in an iterable (like a list, tuple, etc.) and return a map object (an iterator) containing the results.

It allows you to transform data efficiently without using a for loop.

```python

def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]

squared_numbers = map(square, numbers)

print(list(squared_numbers))
```

10. What is the difference between `map()`, `reduce()`, and `filter() functions in Python?

- Map() : Apply a function to every item in an iterable and return a new iterable (map object).

```python
numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)
print(list(result))
```

- Reduce():Repeatedly apply a function to the iterableâ€™s elements to reduce it to one cumulative result.

```python
from functools import reduce
numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)
```

- Filter() : Filter elements from an iterable based on whether the function returns True or False

```python
numbers = [1, 2, 3, 4, 5]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

nums = [10, 15, 20, 25, 30]
print("Sum of even numbers:", sum_of_even_numbers(nums))
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n          # number of terms to generate
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self         # returns the iterator object itself

    def __next__(self):
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b  # update values
            self.count += 1
            return result
        else:
            raise StopIteration  # stops iteration when done

# Example usage
fib = Fibonacci(10)
for num in fib:
    print(num, end=" ")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(limit):
    for i in range(limit + 1):
        yield 2 ** i  # yield generates the next power of 2

# Example usage
for value in powers_of_two(5):
    print(value, end=" ")
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()   # remove newline characters
```
